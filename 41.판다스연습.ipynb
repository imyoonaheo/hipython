#인덱싱

[] 인덱싱

- iloc[] 행기반 인덱싱 >>> [행,열]
- loc[] 이름기반(위치) 인덱싱 >>> [행이름,열이름]

- : >>> start:end까지 슬라이싱
- loc[행이름:행이름]
- iloc[행번호:행번호] >>> end 행번호는 -1까지

```python
# df1[['이름','age']]
df1.loc[ : , ['이름','age']] #loc [행,열]
```

```python
df1.iloc[0:3, 0:2] #행번호, 열
```

```python
df[ df ['score'] > 90 ] #vectorization boolean list의 요소값 True, False
```

```python
cond = df1['age'] >= 27 
```

```python
# [조건 + loc]

# age가 27 이상인 행만 선택하세요.
df1.loc[cond]

# score가 90 이상인 사람의 이름과 score만 출력하세요.
df1.loc[ df1['score'] >= 90, ['이름','score']]

# age가 25 이상이면서 score가 90 이상인 행을 모두 선택하세요.
cond1 = df['age'] >= 25
cond2 = df1['score'] >= 90
df1.loc[cond1 & cond2]
```
