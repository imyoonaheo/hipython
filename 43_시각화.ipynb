# 용도에 맞는 차트 그리기

```python
tips_df = sns.load_dataset('tips')
tips_df
```

# 1변량 그래프: 히스토그램 - 구간, 빈도수 

```python
plt.hist(data=tips_df, x='total_bill', bins=10, rwidth=0.8)
```

```python
#total_bill vs. tip
plt.subplot(1,2,1)
plt.hist(data=tips_df, x='total_bill', bins=10, rwidth=0.8, color='red')

plt.subplot(1,2,2)
plt.hist(data=tips_df, x='tip', bins=10, rwidth=0.8)
```

```python
#1변량 그래프: 박스플롯
```

```python
plt.boxplot(
  [tips_df[tips_df['sex']== 'Female']['total_bill'],
   tips_df[tips_df['sex']== 'Female']['tip']],
  labels=['total_bill/Female', 'tip/Female']
)
plt.grid()
plt.show()
```

```python
thur_lunch_tip = tips_df[
    (tips_df['day'] == 'Thur') & (tips_df['time'] == 'Lunch')
]['tip']

sat_dinner_tip = tips_df[
    (tips_df['day'] == 'Sat') & (tips_df['time'] == 'Dinner')
]['tip']

sun_dinner_tip = tips_df[
    (tips_df['day'] == 'Sun') & (tips_df['time'] == 'Dinner')
]['tip']
```

```python
plt.boxplot(
  [thur_lunch_tip, sat_dinner_tip, sun_dinner_tip],
  labels=['Thurs Lunch', 'Sat Dinner', 'Sun Dinner']
)
plt.show()
```

```python
sizes = sorted(tips_df['size'].unique())
tip_by_size = [tips_df[tips_df['size'] == s]['tip'] for s in sizes]

plt.figure(figsize=(8, 5))
plt.boxplot(tip_by_size, labels=sizes, showfliers=True)

plt.xlabel('Table Size (Number of People)')
plt.ylabel('Tip Amount')
plt.title('Tip Distribution by Table Size')

plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
# 설치가 안 되어 있으면: pip install folium

import folium

# 출발: 노량진역 (대략 좌표)
start = (37.514238, 126.942040)  # Noryangjin Station :contentReference[oaicite:1]{index=1}

# 도착: 사용자 제공 좌표
end = (37.711441, 126.716722)

# 지도 중심(두 점의 중간)
center = ((start[0] + end[0]) / 2, (start[1] + end[1]) / 2)

m = folium.Map(location=center, zoom_start=11)

# 마커
folium.Marker(start, popup="Noryangjin Station", tooltip="Start").add_to(m)
folium.Marker(end, popup="Destination", tooltip="End").add_to(m)

# 선(Polyline)
folium.PolyLine([start, end], weight=5, opacity=0.8).add_to(m)

# 화면에 두 점이 모두 보이도록 자동 확대/이동
m.fit_bounds([start, end])

m  # Jupyter/Colab이면 이 줄로 지도 표시
```
