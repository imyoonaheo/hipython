# 리스트의 특징 

```python
data = list(range(-100,100))
print(len(data))
```

```python
#리스트에 양수만 남긴다
for x in data : 
    if x < 0 :
        data.remove(x) # --> 큰 리스트의 반복 수정은 비용이 큼 → 누적 후 일괄 처리하는 것이 좋습니다.
```

```python
data
```

```python
# for x in data : 
    # if x < 0 :
        # data.remove(x)
filtered = []
for x in data:
    if x >= 0 :
        filtered.append(x)
```
