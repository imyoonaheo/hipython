# 네이버 증권 뉴스 탭의 3일치 기사를 수집

데이터 프레임으로 생성하고 파일로 저장

사이트 주소: https://finance.naver.com/news/mainnews.naver?date=2024-10-25

```python
import pandas as pd # 데이터 정리 및 분석용 라이브러리 (현재 코드에서는 아직 사용되지 않음)
import requests # 웹사이트에 접속해서 데이터를 가져오는 라이브러리
from bs4 import BeautifulSoup # HTML에서 원하는 정보만 추출하기 위한 도구


# 1. 크롤링할 웹사이트 주소 지정
url = 'https://finance.naver.com/news/mainnews.naver?date=2026-02-20'
# headers = {
#    "User-Agent": "Mozilla/5.0"
#}
# 헤더를 설정하지 않으면 차단 당할 수도 있음.

# 2. 해당 웹사이트에 접속 요청 보내기
response = requests.get(url)

# 3. 가져온 HTML 문서를 분석 가능한 형태로 변환
soup = BeautifulSoup(response.text, 'lxml')

# response.text : 웹사이트의 HTML 코드
# 'lxml' : HTML을 빠르게 분석하기 위한 파서(parser)
```

기사제목:

안에 텍스트 본문 URL: 안에 태그 기사 본문: 날짜: URL에서 가져오기

```python
# 기사제목 리스트 가져오기
titles = soup.find_all('dd', {'class':'articleSubject'}) # soup.find_all() → 조건에 맞는 태그를 모두 리스트로 반환

print(titles[:2])
# HTML 구조 확인용 (디버깅 목적)

# 첫번째 기사의 기사 제목
print(titles[0].text.strip())
# .text → 태그 안의 텍스트만 추출
# .strip() → 앞뒤 공백 제거
```

```python
# 첫번째 기사의 상세페이지 URL

# 1. 첫 번째 기사 dd 태그 안에서 a 태그 찾기
titles[0].find('a').get('href') # 2. a 태그의 href 속성 가져오기 (기사 상세 URL)

news_url = 'https://finance.naver.com' + titles[0].find('a').get('href')

# 상세 뉴스페이지 가져오기
news_response = requests.get(news_url)

# 5. 상세 페이지 HTML 파싱
news_soup = BeautifulSoup(news_response.text, 'lxml')

# 6. HTML 구조 확인 (디버깅용)
print(news_soup)
```

```python
#redirect 주소를 반환하고 있으므로, redirecet 주소를 사용하여 다시 상세 뉴스 페이지에 접근

# 1. script 태그 찾기
news_url2 = news_soup.find('script').text.split("'")[1]
# find('script') → 첫 번째 <script> 태그 찾기
# .text → 자바스크립트 코드 문자열
# .split("'")[1] → 작은따옴표 기준으로 나눠서 두 번째 값 추출
# → 실제 이동해야 할 기사 URL 추출

# 상세 뉴스페이지 재도전
# 2. 추출한 실제 기사 URL로 다시 요청
news_response2 = requests.get(news_url2)

# 3. 진짜 기사 HTML 파싱
news_soup = BeautifulSoup(news_response2.text, 'lxml')

# 4. 구조 확인 (디버깅용)
print(news_soup.prettify()[:1000])
```

# 함수로 만들기

```python
#반복문 사용하여 함수로 생성하기
import datetime
import time

def get_news_items(html):
  # 기사제목 리스트 가져오기
  titles = html.find_all('dd', {'class':'articleSubject'})
  title_list = []
  url_list = []
  article_list = []

  for t in titles:
    # 기사의 제목
    title = t.text.strip()

    # 기사의 상세페이지 URL
    news_url = 'https://finance.naver.com' + t.find('a').get('href')

    # 상세페이지 request
    news_response = requests.get(news_url)
    news_soup = BeautifulSoup(news_response.text, 'lxml')

    # 리다이렉트 주소를 파싱
    news_url2 = news_soup.find('script').text.split("'")[1]

    # 상세 뉴스페이지 request 재도전
    news_response2 = requests.get(news_url2)
    news_soup = BeautifulSoup(news_response2.text, 'lxml')

    # 신문기사 본문 파싱
    article = news_soup.find('div', {'class':'newsct_article _article_body'}).text.strip()

    # 리스트에 값 채우기
    title_list.append(title)
    url_list.append(news_url2)
    article_list.append(article)

  df = pd.DataFrame({'기사제목': title_list, '본문url': url_list, '기사본문': article_list})
  return df
```

```python
news_df = pd.DataFrame()

for i in range(3):
    # 오늘 날짜 기준 i일 이전 날짜 구하기
    date= datetime.date.today() - datetime.timedelta(days = i)
    url = f'https://finance.naver.com/news/mainnews.naver?date={date}'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    time.sleep(1)

    # 해당일 신문기사 스크랩 데이터프레임 반환
    temp_df = get_news_items(soup)
    temp_df['date'] = date
    news_df = pd.concat([news_df, temp_df], ignore_index=True).copy()

    print(news_df.head())
    news_df.to_csv('article.csv')
```
