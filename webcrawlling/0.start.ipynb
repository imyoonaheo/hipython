html 속에 있는 요소를 꺼내는 방법 
1. 요소를 직접 지정 > find("요소명", id="id명")
2. CSS의 요소 검색 방법 > select("선택자")

선택자란, CSS에서 요소를 지정하는 방법입니다. 

CSS에서 
- #은 id, .은 클래스를 나타냅니다. "#subect" ".contents"
- 자식은 > "span > b"
- 여러개 요소 목록, "div h1"


- select()는 지정한 선택자의 요소를 모두 가져옵니다. [리스트로]
- select_one()은 지정한 선택자 요소 중 첫번째 요소만 반환합니다. 

```python
soup.select_one("div#subject") #id로 꺼내기
```

```python
soup.select("div.contents") #class로 거내기
```

```python
h1 = soup.select_one("h1")
print(h1.get_text())
```

```python
subject = soup.select_one("#subject") #id로 꺼내기
print(subject.get_text())
```

```python
contents = soup.select_one(".contents")#class로 꺼내기
print(contents.get_text())
```

```python
b_tag = soup.select_one(".contents  b") #contents 클래스 안에 있는 b 태그
print(b_tag.text)
```

```python
b_tag = soup.select("div > b") #div 클래스 안에 있는 b 태그
print(b_tag[0].text)
```

```python
b_tags = soup.select("div b") #div 태그 안에 있는 모든 b 태그

for tag in b_tags:
    print(tag.text)
```

```python
#div 태그의 id가 subject인 태그 안에 있는 b 태그

b_tags = soup.select("div#subject b") 
```

```python
# div 태그의 attr이 contents인 태그 안에 있는 b 태그
b_tags = soup.select("div.contents b")
```

```python
# 1. 타겟 URL 결정
url = "https://finance.naver.com/marketindex/"
response = requests.get(url)

# 3. BeautifulSoup으로 HTML 파싱
soup = BeautifulSoup(response.text, "html.parser")

# 4. USD/KRW 환율 값 찾기
usd_price_elem = soup.select_one(
  "a.head.usd > div.head_info > span.value"
)
usd_price = usd_price_elem.get_text().strip()
print("미국 USD/KRW 환율:", usd_price)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://finance.naver.com/marketindex/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

# JPY/KRW 환율 값 찾기
jpy_price_elem = soup.select_one(
    "a.head.jpy > div.head_info > span.value"
)

jpy_price = jpy_price_elem.get_text().strip()
print("엔화 JPY/KRW 환율:", jpy_price)
```
