```python
import numpy as np
from scipy import stats
```

```python
np.random.seed(42)
```

```python
group_old = np.random.normal(loc=50, scale=10, size=30)
```

```python
group_new = np.random.normal(loc=55, scale=10, size=30)
```

```python
# ë…ë¦½ í‘œë³¸ T-ê²€ì • ìˆ˜í–‰
t_stat, p_value = stats.ttest_ind(
    group_new,       # ì‹ ê·œ ì•½ (ëŒ€ë¦½ê°€ì„¤ì˜ ì£¼ì¸ê³µ)
    group_old,       # ê¸°ì¡´ ì•½
    equal_var=False, # ë“±ë¶„ì‚° ê°€ì • ì•ˆ í•¨ (Welch's t-test, ì‹¤ë¬´ì ìœ¼ë¡œ ë” ì•ˆì „í•¨)
    alternative='greater' # ì¤‘ìš”: "ë” í¬ë‹¤(ìš°ìˆ˜í•˜ë‹¤)"ë¥¼ ê²€ì¦í•˜ë¯€ë¡œ ìš°ì¸¡ ê²€ì •
)
```

```python
from scipy.stats import mannwhitneyu

# ë‘ ê·¸ë£¹ì˜ ì›” ì§€ì¶œì•¡ (ë‹¨ìœ„: ë§Œì›)
high_income = [450, 520, 380, 600, 550, 480, 510, 490, 530, 470]
low_income = [250, 280, 230, 310, 270, 260, 240, 290, 255, 265]

u_stat, p_value = mannwhitneyu(high_income, low_income, alternative='two-sided')
print(f"Uê°’ = {u_stat:.3f}, pê°’ = {p_value:.3f}")
```

```python
from statsmodels.stats.proportion import proportions_ztest

# ì˜ˆì œ: ë‘ ê´‘ê³ ì˜ í´ë¦­ë¥  ë¹„êµ
# ê´‘ê³  A: 100ëª… ì¤‘ 15ëª… í´ë¦­, ê´‘ê³  B: 100ëª… ì¤‘ 25ëª… í´ë¦­
count = np.array([15, 25])  # ì„±ê³µ íšŸìˆ˜
nobs = np.array([100, 100])  # ì „ì²´ ì‹œë„ íšŸìˆ˜

z_stat, p_value = proportions_ztest(count, nobs)
print(f"zê°’ = {z_stat:.3f}, pê°’ = {p_value:.3f}")
```
