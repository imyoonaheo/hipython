```python
# 간식박스 생성 함수
def make_snackbox():
    snackbox = ['두쫀쿠']
    # make_msg = input('먹고 싶은 간식을 추가하세요. 쉼표(,)로 연결하세요').split(',')
    # snackbox += make_msg
    return snackbox
```

```python
# 간식박스 추가함수
def add_snackbox(sbox): #간식 리스트
    make_msg = input('먹고 싶은 간식을 추가하세요. 쉼표(,)로 연결하세요').split(',')
    sbox.extend(make_msg)
    print(sbox)
    return sbox
```

```python
def display_snackbox(sbox):
    print(sbox)
```

```python
#메뉴출력
def display_menu():
    print("1. 간식추가")
    print("2. 간식박스 확인")
    print("3. 간식 제거")
    print("4. 주문확정")
```

```python
def order_confirm():
    make_msg = input('주문을 확정하시겠습니까? (yes/no)')
    if make_msg == 'yes':
        return True
    else : 
        return False
```

```python
def del_menu(sbox):
    make_msg = input('취소할 메뉴를 입력하세요.')
    sbox.remove(make_msg)
    print(sbox)
    return sbox
```

```python
# 전체 보기
# 삭제 함수
# 수정 함수
```

```python
# 서비스의 흐름 로직
snack_box = make_snackbox()
print('간식박스 생성완료', snack_box)

while True :
    display_menu() #
    menu_no = input('메뉴를 선택하세요')
    if menu_no == '1' :
        add_snackbox(snack_box) #간식리스트
    elif menu_no == '2' :
        display_snackbox(snack_box)
    elif menu_no =='3' :
        del_menu(snack_box)
    elif menu_no == '4' :
        display_snackbox(snack_box)
        result = order_confirm()
        if result:
            break
print('고객님의 주문을 감사드립니다')
```
