```python
num1, opr, num2 = input('수식을 입력하세요 (예. 20 + 20)').split()
num1, opr, num2
```

```python
num1 = float(num1)
num2 = float(num2)
num1, num2
```

```python
result = 0 
if opr == '+' : 
    result = num1 + num2
elif opr == '-' :
    result = num1 - num2
elif opr == '*' :
    result = num1 * num2
elif opr == '/' :
    result = num1 / num2
else:
    print('사칙연산을 지원합니다') 
print(f'{num1} {opr} {num2}의 결과는 {result}입니다.')
```

```python
inventory = {
    '노트북':5,
    '마우스':120,
    '키보드':0,
    '모니터':15
}
for name, qty in inventory.items():
    if qty >50:
        status = '충분'
    elif qty >10 :
        status = '적정'
    elif qty >0 :
        status = '부족'
    else : 
        status = '품절'
    print(f' {name}:{qty} 로 {status} 상태입니다.')
```

```python
#가위바위보
import random
player = input('가위바위보 게임 (가위/바위/보)')
computer = random.choice(['가위','바위','보'])

result = '승' if ( (player == '가위' and computer=='보' )or\
                  (player == '바위' and computer=='가위') or\
                  (player == '보' and computer=='바위')      ) else \
        '무승부' if player == computer else\
        '패'
print(f' {player} : {computer} -> {result}')
```
