```python
while True :
    user_input = input('정수의 목록을 공백으로 구분해서 입력하세요. 종료: "q"')
    #종료 명령이면 break
    if user_input == 'q' :
        break
    str_int = user_input.split()
    numbers = []

    #int(user_input)
    for num in str_int:
        # if num == '0' :
        #     continue
        if num != '0':
            numbers.append(int(num))
            
    #입력받은 정수 리스트
    print(numbers)
    #입력한 숫자가 없습니다 체크
    if numbers == None : 
        print('입력한 숫자가 없습니다')
    #양수/음수 판정 
    plus_numbers = []
    minus_numbers = []
    for num in numbers:
        if num > 0 :
            plus_numbers.append(num)
        else :
            minus_numbers.append(num)
    pls_cnt = len(plus_numbers)
    minus_cnt = len(minus_numbers)
    total = sum(numbers)
    # print(plus_numbers, minus_numbers)
    print(f'양수 {pls_cnt}개 음수 {minus_cnt}개 총 합계: {total}')
    print()
```
