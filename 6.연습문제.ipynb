```python
def chk_dup_num(s):
    result = True
    val_list = []
    for num in s: #0123456789
        if num not in val_list: #val_list = 0123456789, 6789012345
            val_list.append(num)
        else: #012322456789(false)
            result = False
    if len(val_list) != 10:
        result = False
    return result

print(chk_dup_num('0123456789'))
print(chk_dup_num('01234'))
print(chk_dup_num('01234567890'))
print(chk_dup_num('6789012345'))
print(chk_dup_num('012322456789'))
```

```python
def convert_int(s):
    #콤마가 있는 숫자를 정수로 변환
    result = s.replace(',','')
    return int(result)

convert_int("1,234,567")
```
