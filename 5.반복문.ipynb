```python
#날짜 입력받기
year = int(input('년도를 입력하시오: '))
month = int(input(' 월을 입력하시오: '))
day = int(input('일을 입력하시오: '))

# print('수요일입니다')
```

```python
import datetime
calc_day = datetime.date(year,month,day).weekday() #요일값 0~ 월~

week_day = ['월', '화', '수', '목', '금', '토', '일']
# week_day[calc_day]
print(f'{week_day[calc_day]}요일입니다')
```

```python
# 총일수 % 7 -> calc_day
#입력일 : 2026-10-5
# 년도 1~2024 년도수 * 365 | 366
# 월: 1~9 월 * 30(4,6,9,11) | 31(1,3,5,7,8,10,12) | 28|29(2-윤년,평년)
#일:일
```

```python
# 연도 날수계산 
#1년 1월 1일부터 해당 날짜까지 지난 총 날짜 수를 구한다 → 그 값을 7로 나눈 나머지로 요일을 판별한다
total_days = 0

for i in range(1,year): #1년부터 2023년까지
    if i % 400 == 0: #윤년
        total_days += 366
    elif i % 100 == 0 : #평년
        total_days += 365
    elif i % 4 == 0 : #윤년
        total_days += 366
    else : #평년
        total_days += 365 

print(f'1st : {total_days}, {year}')

#여기까지는 1년 1월 1일부터 2024년 3월 15일까지 며칠이나 지났지?
# 그 다음은 그 숫자를 7로 나눠서 요일을 정함. 왜냐하면 요일은 7마다 반복되니깐.

# 2024년의 월기준 일수 계산
month_days = [0,31,28,31,30,31,30,31,31,30,31,30,31]
for m in range(1,month):
    total_days += month_days[m]
    
print(f'2nd : {total_days}, {year}')

#올해가 윤년인지 체크 +1 
if month >= 3 :
    #윤년체크
    if year % 400 == 0: #윤년
        total_days += 1
    elif year % 100 == 0 : #평년
        total_days += 0
    elif year % 4 == 0 : #윤년
        total_days += 1
    else : #평년
        total_days += 0 

print(f'3nd : {total_days}, {year}, {month}')       

total_days += day

print(f'final : {total_days}, {year}, {month}, {day}')   

# 연도 + 월 + 일 % 7 > 요일에 해당하는 수
remain = total_days % 7
result = None

if remain == 0:
    result = '일요일'
elif remain == 1:
    result = '월요일'
elif remain == 2:
    result = '화요일'
elif remain == 3:
    result = '수요일'
elif remain == 4:
    result = '목요일'
elif remain == 5:
    result = '금요일'
elif remain == 6:
    result = '토요일'

print(f'{year}년 {month}월 {day}일은 {result}입니다.')
```

# while 문

```python
sum = 0 #sum이라는 변수를 초기화 
for i in range (1,11): #1,2,3 i에 들어있는거임 이런 시퀀스가
    sum += i #처리
    
sum #출력
```

```python
#초기화
i = 1
sum = 0

while True : #조건문
    sum = sum + i
    i = i + 1
    if i >= 10 : #조건식
        break
    # i = i + 1
print(sum, i)
```

```python
user_input = 0
while user_input != "Exit" :
    #True 인동안만 실행
    user_input = input('명령을 입력하세요 (Exit를 넣으면 종료합니다.)')
    print(f'입력하신 명력은 {user_input}입니다.')

print('Exit 명령을 입력하셨습니다.')
```

### 업앤다운 게임

```python
import random
number = random.randint(1,10)
number #컴퓨터의 선택
```

```python
#정답 맞추기
cnt = 0
while True: 
    #입력받기
    answer = int(input('1~10 사이의 숫자를 맞추세요'))
    cnt += 1
    #조건
    if answer == number : #맞춘 경우
        print(f'축하합니다. you got it! 도전횟수: {cnt}')
        break
    elif answer > number :
        print('down')
    else :
        print('up')
```
