- system : 사전에 기본적으로 설정해야하는 프롬프트
- user : 사용자의 입력 
- assistant : 그전에 기억하고 싶은 대화를 저장할 수 있음 -> RAG 로 간다.

```python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("user","2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?"),
        {
            "role": "assistant", #헬퍼는 그대로 두고, assistant만 명시적으로 처리
            "content": [
                {
                    "type": "output_text",
                    "text": "바로 예상을 뚫고 우승한 브라질입니다."
                }
            ]
        },
        msg("user","그 나라가 화제가 되었던 이유를 자세하게 설명해줘")
    ]
)
response.output_text
```

```python
def msg_assist(role, text) :
    msg_src = {"role": role, "content": [{"type":"output_text", "text":text}] }
    return msg_src
```

```python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("user","2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?"),
        msg_assist("assistant","바로 예상을 뚫고 우승한 브라질입니다." ),
        msg("user","그 나라가 화제가 되었던 이유를 자세하게 설명해줘")
    ]
)
response.output_text
```

# 좋은 프롬프트 조건: 역할 + 목표 + 조건 + 출력형식

```python
sys_prompt = """
역할: 너는 B2B 마케팅 전문가야
목표: 인스타그램 랜딩페이지 상단에 넣을 핵심 문구를 작성한다.
조건: 
- 각 문구는 20자 이내로 작성
- 신뢰감과 전문성을 강조
- 과장된 표현과 감탄사 금지

출력형식: 
번호목록 (1~5)
"""
```

```python
sys_prompt_1 = '''
역할: 너는 시사,경제 전문가야.
목표: 뉴스 기사를 넣으면 분석해서 글을 작성한다. 
조건: 
- 입력에 대한 주제를 설명해주세요.
- 기사에 대한 요약을 간단히 해주세요
- 기사에 대해 가능한 질문을 3가지 제안해주세요.

출력형식:
문단을 나눠서 설명해주세요
'''
```

```python
user_content = '''
[데일리안 = 임채현 기자] SK하이닉스가 지난해에 이어 올해에도 성과급 일부를 자사주로 받을 수 있는 '주주 참여 프로그램'을 시행한다.

16일 업계에 따르면 SK하이닉스는 전날 사내 공지를 통해 대표적인 성과급 제도인 초과이익분배금(PS)에 대해 주주 참여 프로그램을 운영한다고 구성원들에게 안내했다.

이 프로그램은 구성원이 PS의 일부를 자사주로 선택해 1년간 보유할 경우, 매입 금액의 15%를 현금으로 추가 지급하는 방식이다. 성과급을 현금 대신 주식으로 받는 선택에 인센티브를 부여해 구성원의 장기적 회사 가치 참여를 유도하겠다는 취지다.

참여를 원하는 구성원은 PS의 최소 10%에서 최대 50%까지 10% 단위로 주식 전환을 선택할 수 있다. 예를 들어 1억 원의 PS를 받는 직원이 최대치인 50%를 주식으로 선택할 경우, 5000만 원 상당의 자사주를 취득하게 되며, 1년 보유 시 이에 대한 15%에 해당하는 현금이 추가 지급된다. 신청 기간은 오는 22일까지다.

SK하이닉스는 구성원의 근로 의욕과 소속감을 높이고, 회사와 구성원 간 이해관계를 일치시키기 위해 해당 제도를 운영하고 있는 것으로 풀이된다. 회사는 지난해에도 동일한 프로그램을 시행한 바 있다.

PS는 연간 실적에 따라 1년에 한 차례 지급되는 성과급으로, 기존에는 연봉의 최대 50%(기본급 1000%) 한도가 적용됐다. 그러나 노사는 지난해 하반기 협의를 통해 지급 한도를 폐지하고, 전년도 영업이익의 10%를 PS 재원으로 삼는 새로운 기준을 도입했다.

이에 따라 PS 산정 금액의 80%는 당해 연도에 지급하고, 나머지 20%는 2년에 걸쳐 매년 10%씩 이연 지급하는 방식이 적용된다. 이번에 지급될 PS부터는 이 같은 개편 기준이 반영된다.

SK하이닉스의 지난해 4분기 실적 발표가 이달 말 예정돼 있는 만큼, PS 지급 시점은 그 이후인 다음 달 초가 될 가능성이 크다. 시장에서는 SK하이닉스의 지난해 연간 영업이익을 약 45조 원으로 전망하고 있으며, 전체 직원 수가 약 3만3000명인 점을 감안하면 단순 계산 기준 1인당 PS 규모는 1억4000만 원 수준에 이를 것으로 추산된다.
'''
```

```python
res = client.responses.create(
    model = 'gpt-4o-mini',
    input = [
        msg("system", sys_prompt_1),
        msg("user", user_content )
        
    ],
    temperature=0
)
res.output_text
```
