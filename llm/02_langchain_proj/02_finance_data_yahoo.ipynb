```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser


prompt = ChatPromptTemplate.from_messages([
  ("system", """Want assistance provided by qualified individuals enabled with experience on understanding charts using technical analysis tools while interpreting macroeconomic environment prevailing across world consequently assisting customers acquire long term advantages requires clear verdicts therefore seeking same through informed predictions written down precisely! First statement contains following content- “Can you tell us what future stock market looks like based upon current conditions ?"."""),
  ("user", """
   {company} 주식에 투자를 해도 될까요? 마크다운 형식의 투자 보고서를 한글로 작성해주세요.
   """)
])


output_parser = StrOutputParser()

chain = prompt | llm | output_parser

company = "Microsoft"

response= chain.invoke({"company": company})
print(response)
```

```python
import yfinance as yf

# Microsoft 주식 심볼 (MSFT)
ticker = 'MSFT'

# 티커 정보 가져오기
msft = yf.Ticker(ticker)

# 기본 정보 조회
basic_info = {
    "회사명": msft.info.get('longName'),
    "산업": msft.info.get('industry'),
    "부문": msft.info.get('sector'),
    "시가총액": msft.info.get('marketCap'),
    "총 발행 주식 수": msft.info.get('sharesOutstanding'),
    "본사 위치": msft.info.get('country'),
    "CEO": msft.info.get('ceo'),  # 일부 주식에서 사용 가능
    "웹사이트": msft.info.get('website'),
}

# 기본 정보 출력
for key, value in basic_info.items():
    print(f"{key}: {value}")
```

```python
import pandas as pd

# 손익계산서 핵심 지표 추출
income = msft.financials
revenue = income.loc["Total Revenue"].iloc[0]
operating_income = income.loc["Operating Income"].iloc[0]
net_income = income.loc["Net Income"].iloc[0]

# 대차대조표 핵심 지표 추출
balance = msft.balance_sheet
total_assets = balance.loc["Total Assets"].iloc[0]
total_debt = balance.loc["Total Debt"].iloc[0]

# 현금흐름표 핵심 지표 추출
cashflow = msft.cashflow
operating_cashflow = cashflow.loc["Operating Cash Flow"].iloc[0]
free_cashflow = cashflow.loc["Free Cash Flow"].iloc[0]

print(f"매출: {revenue:,.0f}")
print(f"영업이익: {operating_income:,.0f}")
print(f"순이익: {net_income:,.0f}")
print(f"총자산: {total_assets:,.0f}")
print(f"총부채: {total_debt:,.0f}")
print(f"영업현금흐름: {operating_cashflow:,.0f}")
print(f"잉여현금흐름: {free_cashflow:,.0f}")
```

```python
# 벨류에이션 지표 추출

info = msft.info

valuation = {
    "PER (주가수익비율)": info.get("trailingPE"),
    "PBR (주가순자산비율)": info.get("priceToBook"),
    "ROE (자기자본이익률)": info.get("returnOnEquity"),
    "EPS (주당순이익)": info.get("trailingEps"),
    "배당수익률": info.get("dividendYield"),
    "52주 최고가": info.get("fiftyTwoWeekHigh"),
    "52주 최저가": info.get("fiftyTwoWeekLow"),
    "현재 주가": info.get("currentPrice"),
}

for key, value in valuation.items():
    print(f"{key}: {value}")
```

```python
# 주가 차트 시각화 

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'  # 한글 폰트
plt.rcParams['axes.unicode_minus'] = False

hist = msft.history(period="6mo")

plt.figure(figsize=(12, 5))
plt.plot(hist.index, hist["Close"], color="royalblue", linewidth=2)
plt.title(f"{ticker} 최근 6개월 주가 추이", fontsize=14)
plt.xlabel("날짜")
plt.ylabel("주가 (USD)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# yfinance 내장 뉴스 기능
news = msft.news

for i, article in enumerate(news[:5]):  # 최근 5개
    content = article.get("content", {})
    print(f"[{i+1}] {content.get('title')}")
    print(f"    출처: {content.get('provider', {}).get('displayName')}")
    print(f"    링크: {content.get('canonicalUrl', {}).get('url')}")
    print()
```

```python
news_summary = ""
for i, article in enumerate(news[:5]):
    content = article.get("content", {})
    title = content.get("title", "")
    summary = content.get("summary", "")  # 요약이 있는 경우
    news_summary += f"{i+1}. {title}\n"
    if summary:
        news_summary += f"   요약: {summary}\n"
    news_summary += "\n"

print(news_summary)
```

```python
import yfinance as yf
import pandas as pd

ticker = 'MSFT'
msft = yf.Ticker(ticker)

# ① 기본 정보
basic_info = {
    "회사명": msft.info.get('longName'),
    "산업": msft.info.get('industry'),
    "부문": msft.info.get('sector'),
    "시가총액": msft.info.get('marketCap'),
    "현재 주가": msft.info.get('currentPrice'),
    "52주 최고가": msft.info.get('fiftyTwoWeekHigh'),
    "52주 최저가": msft.info.get('fiftyTwoWeekLow'),
    "웹사이트": msft.info.get('website'),
}

# ② 밸류에이션 지표
valuation = {
    "PER": msft.info.get('trailingPE'),
    "선행 PER": msft.info.get('forwardPE'),
    "PBR": msft.info.get('priceToBook'),
    "ROE": msft.info.get('returnOnEquity'),
    "ROA": msft.info.get('returnOnAssets'),
    "EPS": msft.info.get('trailingEps'),
    "선행 EPS": msft.info.get('forwardEps'),
    "배당수익률": msft.info.get('dividendYield'),
    "EV/EBITDA": msft.info.get('enterpriseToEbitda'),
    "영업이익률": msft.info.get('operatingMargins'),
    "순이익률": msft.info.get('profitMargins'),
}

# ③ 주가 히스토리
hist_1mo = msft.history(period="1mo")   # 1개월
hist_6mo = msft.history(period="6mo")   # 6개월
hist_1y  = msft.history(period="1y")    # 1년
hist_5y  = msft.history(period="5y")    # 5년

# ④ 재무제표 (연간)
income_statement    = msft.financials       # 손익계산서
balance_sheet       = msft.balance_sheet    # 대차대조표
cashflow_statement  = msft.cashflow         # 현금흐름표

# ⑤ 재무제표 (분기별)
quarterly_income    = msft.quarterly_financials  # 분기 손익계산서
quarterly_balance   = msft.quarterly_balance_sheet
quarterly_cashflow  = msft.quarterly_cashflow

# ⑥ 애널리스트 정보
analyst_targets     = msft.analyst_price_targets    # 목표주가
recommendations     = msft.recommendations_summary  # 매수/중립/매도 의견
earnings_dates      = msft.earnings_dates           # 실적 발표일 및 EPS 예상치

# ⑦ 수급 정보
institutional_holders   = msft.institutional_holders   # 기관 투자자
mutualfund_holders      = msft.mutualfund_holders       # 뮤추얼펀드
insider_transactions    = msft.insider_transactions     # 내부자 거래

# ⑧ 뉴스
news = msft.news

print("✅ 데이터 수집 완료")
print(f"  - 기본 정보: {len(basic_info)}개 항목")
print(f"  - 밸류에이션: {len(valuation)}개 항목")
print(f"  - 주가 히스토리: 1개월({len(hist_1mo)}일) / 6개월({len(hist_6mo)}일) / 1년({len(hist_1y)}일)")
print(f"  - 뉴스: {len(news)}건")
```
