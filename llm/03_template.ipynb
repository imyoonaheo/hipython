```python
from langchain_core.globals import set_llm_cache
from langchain_core.caches import InMemoryCache
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI

set_llm_cache(InMemoryCache())
llm = ChatOpenAI(model="gpt-4o-mini")
```

```python
from langchain_core.prompts  import PromptTemplate

template = PromptTemplate.from_template("{city}에서 {adjective} {topic}을 알려주세요.")
prompt = template.format(city="서울", adjective="가장 유명한", topic="맛집")
print(prompt)

response = llm.invoke(prompt)
print(response.content)
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ("system", 
     "당신은 15년 경력의 CFA 자격증을 보유한 시니어 주식 애널리스트입니다. "
     "재무제표 분석, 기업 밸류에이션, 투자 리스크 평가에 전문성을 갖추고 있으며, "
     "개인 투자자가 이해하기 쉬운 언어로 객관적인 분석을 제공합니다. "
     "단, 투자 판단의 최종 책임은 사용자에게 있음을 항상 명시합니다."),

    ("human", 
     "삼성전자 2024년 4분기 실적을 분석해줘."),

    ("ai",    
     "삼성전자 2024년 4분기 실적 분석입니다.\n\n"
     "📊 핵심 지표\n- 매출: 75.8조 원 (전년 동기 대비 +11%)\n"
     "- 영업이익: 6.5조 원\n- 영업이익률: 8.6%\n\n"
     "✅ 긍정 요인\nHBM 수요 확대로 메모리 단가 상승.\n\n"
     "⚠️ 리스크 요인\n중국 경쟁사 물량 공세, 환율 부담.\n\n"
     "💡 투자 의견\n실적 회복세 긍정적. "
     "※ 본 분석은 참고용이며 투자 판단은 본인 책임입니다."),

    ("human", 
     "{company} {year}년 {quarter}분기 실적을 분석해줘.")
])

chain = chat_template | llm
response = chain.invoke({
    "company": "SK하이닉스",
    "year": "2024",
    "quarter": "3"
})
print(response.content)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "당신은 국내 대형 자산운용사에서 15년간 국내 주식 포트폴리오를 운용한 수석 펀드매니저입니다. "
        "기업의 펀더멘털과 거시경제 흐름을 연결해 분석하며, "
        "PER·PBR·ROE·외국인 수급 등 정량 지표를 근거로 객관적인 시각을 제시합니다. "
        "근거 없는 낙관론과 공포심리를 모두 경계하며, 핵심만 간결하게 답변합니다. "
        "답변 끝에는 반드시 '※ 투자 판단의 최종 책임은 본인에게 있습니다.'를 볼드체로 붙입니다."
        "답변은 항상 단락별로 줄바꿈을 넣어 가독성 있게 작성합니다. "
        "번호 목록이 있을 경우 각 항목 앞에 줄바꿈을 추가합니다. "

    ),
    (
        "user",
        "코스피가 요즘 맨날 떨어지는데 지금 한국 주식 자체를 피해야 하는 건가요?"
    ),
    (
        "ai",
        "하락이 곧 회피 신호는 아닙니다. 먼저 하락의 원인을 구분해야 합니다. "
        "외국인 순매도가 원인이라면 환율·미국 금리 방향을 확인하십시오. "
        "현재 코스피 PBR이 0.9배 이하라면 역사적으로 장기 투자자에게 진입 구간이었습니다. "
        "다만 반도체·2차전지 등 섹터별 실적 모멘텀이 살아있는지 개별 점검이 선행되어야 합니다. "
        "시장 전체를 피하는 것보다 수급이 견조한 섹터를 좁혀가는 전략이 유효합니다. "
    ),
    (
        "human",
        "대상: {target}\n"
        "설명 수준: {level}\n"
        "질문: {question}"
    )
])
```

```python
llm = ChatOpenAI(model="gpt-4o", temperature=0.1)
chain = prompt | llm | StrOutputParser()

response = chain.invoke({
    "target": "삼성전자",
    "level": "일반 투자자",
    "question": "삼성전자가 현재 한국 시장에 미치는 영향이 어느정도인가요?"
})
print(response)
```
