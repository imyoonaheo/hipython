```python
semantic_cache = RedisSemanticCache(
    redis_url=REDIS_URL,
    embeddings=OpenAIEmbeddings(model="text-embedding-3-small"),  # 파라미터명 변경
    distance_threshold=0.02,
    ttl=3600
)
set_llm_cache(semantic_cache)
```

```python
semantic_cache.clear() 
```

```python
messages = [
    ("system", "너는 투자리포트 작가이다."),
    ("human",  "AAPL 최근 분기 핵심 포인트 세 가지를 요약해줘.")
]
resp = llm.invoke(messages)   # 동일 입력이면 Redis에서 바로 HIT
print(resp.content)
```

```python
response = llm.invoke([HumanMessage(content="코스닥에 대해 간단히 설명해줘.")])
print(response.content)
```

```python
import time
# 시간 측정 헬퍼 함수
def ask(question: str) -> tuple:
    start = time.perf_counter()
    response = llm.invoke([HumanMessage(content=question)])
    elapsed = time.perf_counter() - start
    return response.content, elapsed

print("환경 설정 완료")
```

```python
question_pairs = [
    ("코스피 지수란 무엇인가요?",       "최초 API 호출"),
    ("코스피 지수란 무엇인가요?",       "정확 일치 캐시"),
    ("코스피가 뭔지 설명해줘.",         "시맨틱 캐시 (유사 질문 1)"),
    ("KOSPI 지수에 대해 알려주세요.",   "시맨틱 캐시 (유사 질문 2)"),
]

print("[ RedisSemanticCache 테스트 ]")
for q, label in question_pairs:
    answer, elapsed = ask(q)
    print(f"{label}: {elapsed*1000:.0f}ms | 질문: {q}")
print()
print("응답:", answer[:50], "...")
```
