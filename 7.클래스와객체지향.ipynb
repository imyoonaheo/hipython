#### 클래스 - 구조
객체 생성하기 위한 틀
- 객체 구성 요소: 특성(속성) + 행위(메소드)

- 기본 메소드를 상속받아 사용할 수 있다

```python
class Bicycle: #python object 원조부모
  def __init__(self,color,brand):
    self.color = color
    self.brand = brand
  
  def move(self):
    print (f'{self.color} {self.brand} 자전거가 움직입니다.')
    
```

```python
# b1 = Bicycle() #객체생성
# b2 = Bicycle()

b1 = Bicycle('red', '자이언트') #객체생성
b2 = Bicycle('white', '캄파놀로')
b3 = Bicycle('blue', '캄파놀로')
```

```python
Bicycle() #객체생성
```

```python
b1.move()
```

```python
b2.move()
```

```python
class car:
  #1 - 클래스 속성 
  instance_cnt = 0
  
  # def __init__(self): #생성자 정의 
  #   print('자동차 객체가 생성되었습니다')
  def __init__(self, color=''): #생성자 정의
    #2 - 객체의 속성
    self.color = color
    
    car.instance_cnt +=1
    
    print(f'{self.color} 자동차 객체가 생성되었습니다.({car.instance_cnt}대)')
  def auto_pilot(self):
    print(f'{self.color} 자동차가 자율주행중입니다.')
    #00대의 자동차가 운행중입니다. 
    print(f'{car.instance_cnt}대의 자동차가 운행중입니다')
    
  @classmethod
  def count_check(cls):
    print(f'자동차 대수:{cls.instance_cnt}')
```

# 자동차 객체 생성

```python
c1 = car()
c2 = car('red')
```

```python
c1.auto_pilot()
```
